```python
#Independent and Dependent features
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X
```

```python
#Train,Test split
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test= scaler.transform(X_test)
```

# Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
#coefficients and intercept
print(reg.coef_)
```

```python
#Prediction with test data
X_test_predict=reg.predict(X_test)
X_test_predict
```

```python
plt.scatter(y_test,X_test_predict)
```

```python
#Residual
residual = y_test-X_test_predict
```

```python
sns.displot(residual,kind = 'kde')
```

```python
#performance metrics 
from sklearn.metrics import mean_squared_error
from sklearn.metrics import  mean_absolute_error
print(mean_absolute_error(y_test, X_test_predict))
print(mean_squared_error(y_test,X_test_predict))
print(np.sqrt(mean_squared_error(y_test,X_test_predict)))
```

# R square and adjusted R square

```python
#Formula R^2 = 1-SSR/SST
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,X_test_predict)
print(score)
```

#Adjusted R2 = 1-[(1-R^2)*(n-1)/n-k-1)]


Where
R2:The R2 of the model
n: The number of obeservations
k: The number of predictor variables
    

```python
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
